```python
import numpy as np
from ultralytics import YOLO
import cv2

from sort.sort import *
from util import get_car, write_csv, license_complies_format, format_license, read_license_plate
```

```python
model = YOLO("yolov8n.yaml") 
```

```python
# ÐžÐ±ÑƒÑ‡ÐµÐ½Ð¸Ðµ
model = model.train(data='data.yaml', epochs=300, device=0, workers=2) 
```

```python
# Ð—Ð°Ð³Ñ€ÑƒÐ·ÐºÐ° Ð²Ð¸Ð´ÐµÐ¾
cap = cv2.VideoCapture('./sample.mp4')
```

```python
# Ð—Ð°Ð³Ñ€ÑƒÐ·Ð¸Ð¼ Ð¼Ð¾Ð´ÐµÐ»Ð¸ Ð´Ð»Ñ Ð¾Ð¿Ñ€ÐµÐ´ÐµÐ»Ð½Ð¸Ñ Ð¼Ð°ÑˆÐ¸Ð½ Ð¸ Ð½Ð¾Ð¼ÐµÑ€Ð¾Ð²
coco_model = YOLO('yolov8n.pt') # Ð”Ð»Ñ Ð¼Ð°ÑˆÐ¸Ð½
licence_detector = YOLO('./runs/detect/train/weights/last.pt')
auto_tracker = Sort() # Ð¢Ñ€ÐµÐºÐµÑ€ Ð¸Ð· Ð±Ð¸Ð±Ð»Ð¸Ð¾Ñ‚ÐµÐºÐ¸ Sort
```

```python
# ÐŸÐµÑ€ÐµÐ¼ÐµÐ½Ð½Ñ‹Ðµ Ð´Ð»Ñ Ð²Ð²Ð¾Ð´Ð°/Ð²Ñ‹Ð²Ð¾Ð´Ð°
vehicles = [2, 3, 5, 7] # Ð¡Ð¿Ð¸ÑÐ¾Ðº ID Ñ‚Ñ€Ð°Ð½ÑÐ¿Ð¾Ñ€Ñ‚Ð½Ñ‹Ñ… ÑÑ€ÐµÐ´ÑÑ‚Ð², Ñ‡Ñ‚Ð¾Ð±Ñ‹ Ð¼Ð¾Ð´ÐµÐ»ÑŒ Ð½Ðµ Ð¾Ð¿Ñ€ÐµÐ´ÐµÐ»ÑÐ»Ð° Ð²ÑÑ‘, Ñ‡ÐµÐ¼Ñƒ Ð¾Ð±ÑƒÑ‡ÐµÐ½Ð°
frame_num = -1
results = {}
```

```python
# Ð§Ñ‚ÐµÐ½Ð¸Ðµ Ð²Ð¸Ð´ÐµÐ¾
ret = True
while ret:
    frame_num += 1
    ret, frame = cap.read()
    if ret:
        results[frame_num] = {}
        # ÐžÐ¿Ñ€ÐµÐ´ÐµÐ»ÑÐµÐ¼ Ð°Ð²Ñ‚Ð¾
        auto_detections = coco_model(frame)[0]
        detections_ = [] # Ð¥Ñ€Ð°Ð½Ð¸Ñ‚ Ð¾Ð¿Ñ€ÐµÐ´ÐµÐ»Ð½Ð¸Ñ Ð²ÑÐµÑ… Ð°Ð²Ñ‚Ð¾
        for detection in auto_detections.boxes.data.tolist():
            x1, y1, x2, y2, score, class_id = detection
            if int(class_id) in vehicles:
                detections_.append([x1, y1, x2, y2, score])

        # ÐžÑ‚ÑÐ»ÐµÐ¶Ð¸Ð²Ð°Ð½Ð¸Ðµ Ð¼Ð°ÑˆÐ¸Ð½
        track_ids = auto_tracker.update(np.asarray(detections_)) # Ð¥Ñ€Ð°Ð½Ð¸Ñ‚ bboxs Ð²ÑÐµÑ… Ð¾Ð¿Ñ€ÐµÐ´ÐµÐ»Ñ‘Ð½Ð½Ñ‹Ñ… Ð¼Ð°ÑˆÐ¸Ð½

        # ÐžÐ¿Ñ€ÐµÐ´ÐµÐ»ÐµÐ½Ð¸Ðµ Ð½Ð¾Ð¼ÐµÑ€Ð¾Ð²
        plates_detections = licence_detector(frame)[0]
        for plate in plates_detections.boxes.data.tolist():
            x1, y1, x2, y2, score, class_id = plate

            # ÐŸÑ€Ð¸Ð²ÑÐ·ÐºÐ° Ð½Ð¾Ð¼ÐµÑ€Ð° Ðº ÐºÐ¾Ð½ÐºÑ€ÐµÑ‚Ð½Ð¾Ð¹ Ð¼Ð°ÑˆÐ¸Ð½Ðµ
            xcar1, ycar1, xcar2, ycar2, car_id = get_car(plate, track_ids)

            if car_id != -1:
                # Ð’Ñ‹Ð´ÐµÐ»ÐµÐ½Ð¸Ðµ Ð½Ð¾Ð¼ÐµÑ€Ð¾Ð²
                plate_crop = frame[int(y1):int(y2), int(x1):int(x2), :]

                # Ð Ð°ÑÐ¿Ð¾Ð·Ð½Ð¾Ð²Ð°Ð½Ð¸Ðµ ÑÐ¸Ð¼Ð²Ð¾Ð»Ð¾Ð² Ð½Ð° Ð½Ð¾Ð¼ÐµÑ€Ð°Ñ…
                # Ð‘Ð¸Ð½Ð°Ñ€Ð¸Ð·Ð°Ñ†Ð¸Ñ Ð½Ð¾Ð¼ÐµÑ€Ð¾Ð²
                gray_plate_crop = cv2.cvtColor(plate_crop, cv2.COLOR_BGR2GRAY)
                _, thresh_plate_crop = cv2.threshold(gray_plate_crop, 64, 255, cv2.THRESH_BINARY_INV) # ÐÐ¸Ð¶Ðµ 64 Ð²ÑÐµ Ð¿Ð¸ÐºÑÐµÐ»Ð¸ Ð¼Ð°ÐºÑÑÑ‚ÑÑ Ð´Ð¾ 255, Ð½Ð¸Ð¶Ðµ Ð´Ð¾ 0
                plate_text, plate_text_conf_score = read_license_plate(thresh_plate_crop)

                if plate_text is not None:
                    results[frame_num][car_id] = {'car': {'bbox': [xcar1, ycar1, xcar2, ycar2]}, 
                                                'license_plate': {'bbox': [x1, y1, x2, y2], 
                                                                'text': plate_text, 
                                                                "bbox_score": score, 
                                                                'text_score': plate_text_conf_score}}
```

```python
# Ð’Ñ‹Ð²Ð¾Ð´ Ñ‚ÐµÐºÑÑ‚Ð°
write_csv(results, './test.csv')
```
